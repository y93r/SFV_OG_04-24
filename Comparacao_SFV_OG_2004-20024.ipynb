# Comparação entre o perfil das convocadas para os jogos olímpicos 2004 a 2024

Zé Roberto Guimarães assumiu a seleção feminina de vôlei em 2003 e tem sido o treinador desde então, tendo participado das Olimpíadas de [Atenas 2004](https://women.volleybox.net/pt/women-the-olympics-2004-o612), [Pequim 2008](https://women.volleybox.net/pt/women-the-olympics-2008-o611), [Londres 2012](https://women.volleybox.net/pt/women-the-olympics-2012-o610), [Rio 2016](https://women.volleybox.net/pt/women-the-olympics-2016-o965), [Tóquio 2020](https://women.volleybox.net/pt/women-the-olympics-2021-o8363) e [Paris 2024](https://women.volleybox.net/pt/women-the-olympics-2024-o30223/classification). Ele, que já havia conquistado o ouro inédito com a seleção masculina em Barcelona 1992, conseguiu o mesmo feito com a feminina em Pequim 2008 e repetiu o feito em Londres 2012. No vôlei, ele é o primeiro treinador a ser campeão nos dois naipes e o primeiro tri-campeão olímpico do Brasil.

Este é o retrospecto dele com a seleção feminina nas Olimpíadas:



| Olimpíadas              | Posição                                               |
| ----------------- | ---------------------------------------------------------------- |
| Atenas 2004      | 4º Lugar|
| Pequim 2008     | 1º Lugar |
| Londres 2012      | 1º Lugar |
| Rio 2016       | 5º Lugar |
| Tóquio 2020      | 2º Lugar |
| Paris 2024     | 3º Lugar |

## Carregamento dos dados

```python
#bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
og_24 = pd.read_csv(r'olimpiadas2024.csv',sep= ',',encoding='utf-8')
og_20 = pd.read_csv(r'olimpiadas2020.csv',sep= ',',encoding='utf-8')
og_16 = pd.read_csv(r'olimpiadas2016.csv',sep= ',',encoding='utf-8')
og_12 = pd.read_csv(r'olimpiadas2012.csv',sep= ',',encoding='utf-8')
og_08 = pd.read_csv(r'olimpiadas2008.csv',sep= ',',encoding='utf-8')
og_04 = pd.read_csv(r'olimpiadas2004.csv',sep= ',',encoding='utf-8')

display(og_24)
display(og_20)
display(og_16)
display(og_12)
display(og_08)
display(og_04)
```

```python
#infos
og_24.info()
og_20.info()
og_16.info()
og_12.info()
og_08.info()
og_04.info()
```

## Limpeza dos dados

```python
#remover espaço do cabeçalho das colunas
og_24.columns = og_24.columns.str.lstrip()
og_24.columns = og_24.columns.str.replace(' ', '')

og_20.columns = og_20.columns.str.lstrip()
og_20.columns = og_20.columns.str.replace(' ', '')

og_16.columns = og_16.columns.str.lstrip()
og_16.columns = og_16.columns.str.replace(' ', '')

og_12.columns = og_12.columns.str.lstrip()
og_12.columns = og_12.columns.str.replace(' ', '')

og_08.columns = og_08.columns.str.lstrip()
og_08.columns = og_08.columns.str.replace(' ', '')

og_04.columns = og_04.columns.str.lstrip()
og_04.columns = og_04.columns.str.replace(' ', '')
```

```python
#antes de concatenar tirar colunas desnecessárias
og_24 = og_24.drop(columns=['Nº'])
og_20 = og_20.drop(columns=['Nº'])
og_16 = og_16.drop(columns=['Nº'])
og_12 = og_12.drop(columns=['Nº'])
og_08 = og_08.drop(columns=['Nº'])
og_04 = og_04.drop(columns=['Nº'])
```

```python
#removendo o espaço das colunas a coluna Ano Nasc, passa a ser AnoNasc
og_24 = og_24.drop(columns=['AnoNasc'])
og_20 = og_20.drop(columns=['AnoNasc'])
og_16 = og_16.drop(columns=['AnoNasc'])
og_12 = og_12.drop(columns=['AnoNasc'])
og_08 = og_08.drop(columns=['AnoNasc'])
og_04 = og_04.drop(columns=['AnoNasc'])
```

```python
#antes de concatenar, criar coluna da Olimpiada jogada
og_04['Olimpiadas'] = 'Atenas 2004'
og_08['Olimpiadas'] = 'Pequim 2008'
og_12['Olimpiadas'] = 'Londres 2012'
og_16['Olimpiadas'] = 'Rio 2016'
og_20['Olimpiadas'] = 'Tóquio 2020'
og_24['Olimpiadas'] = 'Paris 2024'
```

```python
#concatenar as tabelas
df = pd.concat([og_04,og_08,og_12,og_16,og_20,og_24])
df = df.reset_index(drop=True)
df_group = df.groupby(list(df.columns))
display(df)
```

```python
#Estatistica do df
df.describe()
```

```python
#qtde por posição
df['Posição'].value_counts()
```

## Análise de idade e altura

```python
#comparativo média de idade e altura | Time com maior média de idade
df.groupby('Olimpiadas')[['Idade','Altura']].mean().round(2).sort_values('Idade', ascending=False)
```

## Visualização Gráfica de Idade e Altura

```python
#Média Idade por Olimpiadas
sns.set(style='darkgrid')  #Estilo do gráfico

df.groupby('Olimpiadas')['Idade'].mean()
media_idade = sns.barplot(x='Olimpiadas', y='Idade', data=df, errorbar=None, palette='crest')
media_idade.set_ylim(22,32)
media_idade.set_title('Média Idade por Olimpiadas', fontsize = 16)
media_idade.set_xticklabels(media_idade.get_xticklabels(), horizontalalignment='center', fontsize=8)
plt.show()
```

```python
#Distribuição da variável Idade
idade = df['Idade']
sns.histplot(data=df, x=idade, kde=True, color='DarkGreen').set_title(f"Distribuição da variável Idade", fontsize = 16)
plt.show()
```

```python
#Distribuição da variável Idade por Olimpíada

#Criar um FacetGrid para separar os histogramas por Olimpíada
warnings.filterwarnings('ignore', category=UserWarning, module='seaborn.axisgrid') #ignorar avisos
g = sns.FacetGrid(df, col='Olimpiadas', col_wrap=3, sharex=False, sharey=False)
g.map(sns.histplot, 'Idade', kde=True, color='DarkGreen')

#Adicionar títulos aos subplots
for ax in g.axes.flat:
    ax.set_title(ax.get_title().replace('Olimpiadas = ', ''))

#Adicionar título principal ao gráfico
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Distribuição da variável Idade por Olimpíada', fontsize=16)

#Exibir o gráfico
plt.show()
```

```python
#comparativo média de  altura e idade| Time com maior média de altura
df.groupby('Olimpiadas')[['Altura','Idade']].mean().sort_values('Altura', ascending=False)
```

```python
#Média Altura por Olimpiadas
df.groupby('Olimpiadas')['Altura'].mean()
media_idade = sns.barplot(x='Olimpiadas', y='Altura', data=df, errorbar=None, palette='crest')
media_idade.set_ylim(1.78,1.86)
media_idade.set_title('Média Altura por Olimpiadas', fontsize = 16)
media_idade.set_xticklabels(media_idade.get_xticklabels(), horizontalalignment='center', fontsize=8)
plt.show()
```

```python
#Distribuição da variável Altura
altura = df['Altura']
sns.histplot(data=df, x=altura , kde=True, color='DarkGreen').set_title(f"Distribuição da variável Altura", fontsize = 16)
plt.show()
```

```python
#Criar um FacetGrid para separar os histogramas por Olimpíada
g = sns.FacetGrid(df, col='Olimpiadas', col_wrap=3, sharex=False, sharey=False)
g.map(sns.histplot, 'Altura', kde=True, color='DarkGreen')

#Adicionar títulos aos subplots
for ax in g.axes.flat:
    ax.set_title(ax.get_title().replace('Olimpiadas = ', ''))

#Adicionar título principal ao gráfico
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Distribuição da variável Altura por Olimpíada', fontsize=16)

#Exibir o gráfico
plt.show()
```

```python
#média de altura somente atacantes
atacantes = df.loc[df['Posição'] != 'Líbero']
media_atacantes = atacantes.groupby('Olimpiadas')['Altura'].mean().sort_values(ascending=False)

media_atacantes
```

```python
#plt.figure(figsize=(10, 6))
media_atacantes_og = sns.barplot(x=media_atacantes.index, y=media_atacantes.values, palette='crest')
media_atacantes_og.set_ylim(1.78,1.90)
media_atacantes_og.set_title('Média de Altura das Atacantes por Olimpíada', fontsize=16)
media_atacantes_og.set_xlabel('Olimpíadas')
media_atacantes_og.set_ylabel('Média de Altura')
media_atacantes_og.set_xticklabels(media_atacantes_og.get_xticklabels(), horizontalalignment='center', fontsize=8)
plt.show()
```

```python
#Jogadoras com mais de 1.90
jogadoras_altas = df.loc[df['Altura'] >= 1.90]
as_mais_altas = jogadoras_altas.groupby('Olimpiadas')['Altura'].count().sort_values(ascending=False)

as_mais_altas
```

```python
#Analise por posição e idade
analise_posicao_idade = df.groupby( by=['Posição'] ).agg(
    Quantidade = ('Idade', 'count'),
    Media = ('Idade', 'mean'),
    Max = ('Idade', 'max'),
    Min = ('Idade', 'min'),
    Mediana = ('Idade', 'median'),
    DesvioPadrao = ('Idade', 'std'),
    Variacao = ('Idade', 'var')
)


analise_posicao_idade.sort_values('Media', ascending=False ).round(2)
```

```python
#Analise por posição e altura
analise_posicao_altura = df.groupby( by=['Posição'] ).agg(
    Quantidade = ('Altura', 'count'),
    Media = ('Altura', 'mean'),
    Max = ('Altura', 'max'),
    Min = ('Altura', 'min'),
    Mediana = ('Altura', 'median'),
    DesvioPadrao = ('Altura', 'std'),
    Variacao = ('Altura', 'var')
)


analise_posicao_altura.sort_values('Media', ascending=False ).round(2)
```

```python
#Média de idade e altura por olimpiadas e posição
agrupar = df.groupby(["Olimpiadas", "Posição"], as_index=True)
colunas_numericas = df.select_dtypes(include=[np.number]).columns
resultado = agrupar[colunas_numericas].aggregate(np.mean).round(2).reset_index()
resultado['Ano'] = resultado['Olimpiadas'].apply(lambda x: x.split()[-1])
display(resultado)
```

```python
#Transformar a coluna Ano em numerica
resultado['Ano'] = pd.to_numeric(resultado['Ano'])
```

```python
resultado.info()
```

```python
#Ordenar o DataFrame resultado pelo ano
resultado = resultado.sort_values(by='Ano')

#Criar gráficos para média de idade e altura
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

#Gráfico para média de idade
sns.barplot(x='Ano', y='Idade', hue='Posição', data=resultado, ax=axes[0])
axes[0].set_title('Média de Idade por Olimpiada e Posição', fontsize=16)
axes[0].set_ylim(20, 40)  
axes[0].legend(title='Posição', bbox_to_anchor=(0.5, 1.01), loc='upper center', ncol=5)

#Gráfico para média de altura
sns.barplot(x='Ano', y='Altura', hue='Posição', data=resultado, ax=axes[1])
axes[1].set_title('Média de Altura por Olimpiada e Posição', fontsize=16)
axes[1].set_ylim(1.50, 2.0)  
axes[1].legend(title='Posição', bbox_to_anchor=(0.5, 1.01), loc='upper center', ncol=5)

#Ajustar o layout
plt.tight_layout()

#Exibir os gráficos
plt.show()
```

## Análise Faixa Etária

```python
#criando coluna para mostrar novatas, experientes e veteranas (novatas <= 25, esperientes entre 26 e 33, veteranas =>34)
import sys

df['FaixaEtaria']=pd.cut(
    df['Idade'],
    bins=[0, 25, 33, sys.maxsize],
    labels=['Novata', 'Experiente', 'Veterana']

)

df
```

```python
#Qtde por faixa etária
df['FaixaEtaria'].value_counts()
```

```python
#Visualização gráfica
df['FaixaEtaria'].value_counts().plot(kind='pie',autopct='%1.1f%%', startangle=190, shadow=True)
plt.legend(df['FaixaEtaria'].value_counts(), title="Qtde em nº",  bbox_to_anchor=(1, 1))
plt.axis('equal')
plt.tight_layout()
```

## Análise por participação olímpica

```python
#Jogadoras com mais particapações sob o comando do ZRG
jogadoras = df.groupby(['Nome']).size().reset_index(name='ParticipaçãoOlimpíadas')
jogadoras_3_ou_mais = jogadoras[jogadoras['ParticipaçãoOlimpíadas'] >= 3]
jogadoras_3_ou_mais = jogadoras_3_ou_mais.sort_values(by='ParticipaçãoOlimpíadas', ascending=False)
display(jogadoras_3_ou_mais)
```

```python
#Relação entre faixa etária e particpações em olimpiadas
df.groupby('FaixaEtaria')['ParticipaçãoOlimpíadas'].value_counts()
```

```python
#particapações em olimpiadas
part_og = df['ParticipaçãoOlimpíadas'].value_counts()
print(part_og)
```

```python
#Adicionar uma coluna para a classificação de participação
df['ClassificacaoParticipacao'] = df['ParticipaçãoOlimpíadas'].apply(lambda x: 'Com Participação' if x != 'Estreante' else x)
display(df[['Nome', 'Olimpiadas', 'ParticipaçãoOlimpíadas', 'ClassificacaoParticipacao']].head())
```

```python
#Classificação de participação por Olimpíada
participacao_por_olimpiada = df.groupby(['Olimpiadas', 'ClassificacaoParticipacao']).size().reset_index(name='Contagem')
total_por_olimpiada = participacao_por_olimpiada.groupby('Olimpiadas')['Contagem'].transform('sum')
participacao_por_olimpiada['Porcentagem'] = (participacao_por_olimpiada['Contagem'] / total_por_olimpiada * 100).round(2)
display(participacao_por_olimpiada)
```
